# **1: Top 10 highest rated iphone on flipkart in INDIA.**

## Acc to question we have to find the highest rating iphone.So,we check **Product** and **Star rating** from the data





```python
# we have to find highest rating so we have to sort the data in decresing order

highest_rated = data.sort_values(by=['Star Rating'],ascending = False)

highest_rated = highest_rated.head(10)
```

```python
print(highest_rated['Product Name'])
```

```python
print(highest_rated)
```

# **2: How many rating the do the highest rated iphone on flipkart have**

```python
# so we need Product name and numbers of rating from the data

iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Ratings']
figure = px.bar(highest_rated, x=label , y = counts,title = 'Number of Rating of Highest rated iphone')
figure.show()
```

# **3: Which iphone has the highest number of review on flipkart**

```python
data.head()
```

```python


iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Reviews']
figure = px.bar(highest_rated, x=label , y = counts,title = 'Number of Reviews of Highest rated iphone')
figure.show()
```

#**4: what is the relationship between the sales price of iphone and no of rating of iphone in flipkart**

```python
# it asking relation ship so we use scattered plot graph
```

```python
figure = px.scatter(data_frame= data , x = 'Number Of Ratings',
           y='Sale Price', size = 'Discount Percentage',
           trendline = 'ols',title = 'Relationship between sales price and number of rating')
figure.show()
```

# **5: Relationship between Discount percentage and number of rating**

```python
figure = px.scatter(data_frame= data , x= 'Number Of Ratings',
           y='Discount Percentage',size = 'Sale Price',
           trendline = 'ols', title = 'Relationship between Discount percentage and number of Rating')
figure.show()
```

# **6: Figure out the most least and most expensive iphone in the indian market , along with all their specification**

```python
# so we talking about expensive so we usd Sale price from dataset
```

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

print("Most expensive Model: ")
print(most_expensive)

print("\nLeast expensive Model: ")
print(least_expensive)
```
